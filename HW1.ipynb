```python
# 121. Best Time to Buy and Sell Stock
def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        if len(prices)<2:
            return 0
        profit=0
        min_price=prices[0]
        for i in range(1,len(prices)):
            if prices[i]<min_price:
                min_price=prices[i]
            if prices[i]-min_price>profit:
                profit=prices[i]-min_price        
        return profit
```

```python
# 448. Find All Numbers Disappeared in an Array

def findDisappearedNumbers(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        res=[]
        for val in nums:
            nums[abs(val)-1]=-abs(nums[abs(val)-1])
        for i in range(len(nums)):
            if nums[i]>0:
                res.append(i+1)         
        return res
```

```python
# 15. 3sum
def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        sol = []
        nums.sort()
        for i in range(len(nums)-2):
            if (i>0) and (nums[i-1]==nums[i]):
                continue
            j = i+1
            k = len(nums)-1
            while(j<k):
                temp = nums[j]+nums[k] + nums[i]
                if temp == 0:
                    sol.append([nums[i],nums[j],nums[k]])
                    while (j<k) and (nums[j]==nums[j+1]):
                        j+=1
                    while (j<k) and (nums[k]==nums[k-1]):
                        k-=1
                    j += 1
                    k -= 1
                elif temp < 0:
                    j+=1
                else:
                    k-=1
        return sol
```

```python
# 16. 3sum closest

def threeSumClosest(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        sol, diff = 0, float('inf')
        nums.sort()
        for i in range(len(nums)-2):
            if (i>0) and (nums[i-1]==nums[i]):
                continue
            l,r = i+1, len(nums)-1
            while (l<r):
                temp = nums[i]+nums[l]+nums[r] - target
                if abs(temp)<diff:
                    sol = temp + target
                    diff = abs(temp)
                if temp == 0:
                    return sol
                elif temp > 0:
                    r -= 1
                else:
                    l += 1
        return sol
```

```python
18. 4sum
def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        if len(nums)<4:
            return []
        nums.sort()
        res=set()
        for i,a in enumerate(nums[:-3]):
            if i>1 and a==nums[i-1]:
                continue
            for j, b in enumerate(nums[i+1:-2]):
                if j>1 and b==nums[j-1]:
                    continue
                check=set()
                for c in nums[i+j+2:]:
                    if c not in check:
                        check.add(target-a-b-c)
                    else:
                        res.add((a,b,c,target-a-b-c))
                        check.remove(c)
        return list(map(list,res))
            
```
